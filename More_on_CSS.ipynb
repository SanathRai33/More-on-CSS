7. Create a simple webpage with a card similar to question number 2. The card must be semi-transparent as default and on hovering the card must be displayed clearly. Use CSS opacity to achieve this output.

Ans -->

<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Centered Image and Text</title>
    <style>
        body{
            padding: 0;
            margin: 0;
            box-sizing: border-box;
        }
        .card{
            text-align: center;
            background-color: rgb(255, 170, 0);
            border-radius: 10px;
            padding: 10px;
            width: 250px;
            display: inline-block;
            opacity: 0.5;
            margin: 20px auto;
            transition: opacity 0.5s ease-in-out;

        .card:hover {
            opacity: 1; /* Fully visible on hover */
        }

            width: 250px;
            border-radius: 8px;
            width: 100%;
            display: block;
        }
    </style>
</head>
<body>
    <div class="card">
